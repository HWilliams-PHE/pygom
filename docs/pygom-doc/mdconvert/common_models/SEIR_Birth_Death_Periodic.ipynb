# `.SEIR_Birth_Death_Periodic`

Now extending the SEIR to also have periodic contact, as in
[\[Aron1984\]]().

$$\begin{aligned}
\frac{dS}{dt} &= \mu - \beta(t)SI - \mu S \\
\frac{dE}{dt} &= \beta(t)SI - (\mu + \alpha) E \\
\frac{dI}{dt} &= \alpha E - (\mu + \gamma) I \\
\frac{dR}{dt} &= \gamma I.
\end{aligned}$$
